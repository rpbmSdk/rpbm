```python
import requests
session = requests.Session()    
```

```python
headers = {
    "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7",
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/132.0.0.0 Safari/537.36"
}
session.headers.update(headers)
```

```python
def auth():

    r = session.get(VSF_LOGIN_URL)
    page = bs.BeautifulSoup(r.text, 'html.parser')
    _token = page.find('input', {'name': '_token'})['value']
    r = session.post(VSF_LOGIN_URL, data={
        'login': VSF_LOGIN,
        'password': VSF_PASSWORD,
        '_token': _token,
        'customer_id': None
    })
    # r.raise_for_status()
    return r

r = auth()
r.status_code
```

```python
VSF_SEARCH_URL = 'https://client.myvsf.fr/catalogue/vitrage'
def searchEurocode(eurocode:str='6539RGSH5RD'):
    return session.get(VSF_SEARCH_URL, params={
        'search': eurocode
    })

r = searchEurocode()
# print(r.text)
page = bs.BeautifulSoup(r.text, 'html.parser')
catalog = page.find('div', {'id': 'catalog'})
print(catalog)
```
